```python
import pandas as pd
import numpy as np
```

```python
# Downloading the clumsy dataset from my google drive()
df = pd.read_csv('/content/drive/MyDrive/hh_ru.csv', delimiter=";")
df.head()
```

```python
df.shape
```

```python
# Check for empty values
df = df.replace('..', np.nan)
total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum() / df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.reset_index(inplace=True)
missing_data.rename(columns={'index': 'Attribute'}, inplace=True)
missing_data.head(10)
```

```python
# Deletion of rows with empty values
df.dropna(inplace=True)
```

```python
# Recheck
df = df.replace('..', np.nan)
total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum() / df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.reset_index(inplace=True)
missing_data.rename(columns={'index': 'Attribute'}, inplace=True)
missing_data.head(10)
```

Here are some transfromations of data for easier further work

```python
df['experience'] = df['experience'].apply(lambda x: int(str(x).split()[2]))
```

```python
df['salary_bottom'] = df['salary_bottom'].apply(lambda x: int(str(x).split(' ')[0]))
```

```python
df['education'] = df['education'].apply(lambda x: x.split(' ')[:1])
```
